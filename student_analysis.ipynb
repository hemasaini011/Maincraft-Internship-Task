```python
#  Import Libraries
## We import necessary Python libraries for data analysis and visualization.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Load Dataset
## We load the student dataset using pandas.
df = pd.read_csv('C:/Users/DELL/Downloads/student/student-mat.csv', sep=';')
df.head()
```

```python
#  Data Exploration
## We check the structure, size, and basic statistics of the dataset.
df.shape
df.describe()
```

```python
#  Data Cleaning
## We check for missing values and remove duplicate records.
df.isnull().sum()
df.drop_duplicates(inplace=True)
```

```python
# Data Analysis
## We perform analysis to answer important questions about student performance.
### Average Final Grade
(df['G3']).mean()
```

```python
## Students Scoring Above 15
(df['G3'] > 15).sum()
```

```python
## Correlation Between Study Time and Performance
df[['studytime', 'G3']].corr()
```

```python
## Gender-wise Performance
df.groupby('sex')['G3'].mean()
```

```python
#  Data Visualization
## We visualize the data using graphs for better understanding.
### Grade Distribution
plt.hist(df['G3'])
plt.title("Grade Distribution")
plt.show()
```

```python
## Study Time vs Grades
sns.scatterplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Grades")
plt.show()
```

```python
## Average Score by Gender
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Average Score by Gender")
plt.show()

```
